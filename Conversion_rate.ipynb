<font color = "navy"> <span style = "font-family : 'Times New Roman' ; font-size: 16pt" > <b> III-2.&nbsp;&nbsp;&nbsp; <b> Analyse Bivariee :</b>  </span></font>

````python
# Définir les variables catégorielles
cat_vars = ['country', 'new_user', 'source', 'converted']

# Définir une palette de couleurs personnalisée
custom_colors = ['royalblue', 'darkturquoise']

# Créer une figure avec sous-graphiques
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,12), dpi=200)

# Boucle sur chaque variable catégorielle
for i, var in enumerate(cat_vars):
    row = i // 2
    col = i % 2
    
    # Calculer les proportions de conversion pour chaque catégorie
    conversion_prop = (data.groupby([var, 'converted'])['converted'].count() / data.groupby([var])[var].count()).unstack(level=1)
    
    # Plot stacked bar charts
    conversion_prop.plot(kind='bar', ax=ax[row, col], stacked=True, width=0.8, color=custom_colors, legend=False)
    
    # Add counts and proportions to bars
    for idx, val in enumerate(conversion_prop.index):
        proportion = conversion_prop.loc[val, 1]
        count = data[data[var] == val]['converted'].sum()
        y_location = conversion_prop.loc[val].cumsum()[1] - proportion / 2
        
        ax[row, col].text(x=idx-0.22, y=y_location, s=f'{count}\n({proportion:.1%})', color='black', fontsize=9, fontweight='bold', ha='left')
    
    # Add legend
    ax[row, col].legend(title='Conversion', loc='upper right', fontsize=8)
    
    # Set y limit
    ax[row, col].set_ylim([0, 1.12])
    
    # Rotate xticks
    ax[row, col].set_xticklabels(ax[row, col].get_xticklabels(), rotation=0)
    
    # Set titles and labels
    ax[row, col].set_title(var)
    ax[row, col].set_xlabel('Modalités')
    ax[row, col].set_ylabel('Proportion des converted (%)')

plt.suptitle('Categorical Features vs Target Stacked Barplots', fontsize=22)
plt.tight_layout()    

plt.savefig('dist_biv.jpg')
plt.show()
````

Un ensemble de données déséquilibré est un ensemble de données dont les classes ne sont pas équilibrées ou égales. Cela signifie qu'il y a un nombre inégal d'échantillons de chaque classe cible et que certaines classes peuvent avoir beaucoup plus d'échantillons que d'autres.

<h2 align="left"><font color='royalblue'> L'approche model evaluation:</font></h2>

L'approche pour __l'évaluation du modèle__ lorsqu'il s'agit d'__ensembles de données déséquilibrés__ consiste à utiliser des mesures appropriées, notamment __le rappel, la précision, le score F1 et l'AUC (aire sous la courbe)__ qui conviennent mieux lorsqu'il s'agit d'ensembles de données déséquilibrés, au lieu des mesures de précision traditionnelles qui ne prennent pas en compte le déséquilibre de la classe par défaut.

Dans ce projet, l'objectif principal est de classer les visiteurs potentiels qui sont plus susceptibles de converter. Les mesures utilisées pour évaluer les performances du modèle seront importantes pour déterminer dans quelle mesure le modèle est capable d'identifier ces visiteurs potentiels.

* Le rappel est une mesure de la proportion de cas positifs réels qui ont été correctement identifiés par le modèle. Un score de rappel élevé signifie que le modèle a un faible nombre de faux négatifs, ce qui est souhaitable dans le cadre de ce projet car cela signifie que le modèle ne manque pas de nombreux visiteurs potentiels.

* La précision est une mesure de la proportion de cas positifs identifiés par le modèle qui sont réellement positifs. Un score de précision élevé signifie que le modèle a un faible nombre de faux positifs, ce qui est souhaitable dans ce projet car cela signifie que le modèle n'identifie pas de nombreux visiteurs non convertus comme des visiteurs convertus.

* Le score F1 est une mesure du compromis entre le rappel et la précision. Il est calculé comme la moyenne harmonique du rappel et de la précision. Un score F1 élevé indique un équilibre entre un rappel élevé et une précision élevée.


Pour ce projet, le rappel et la précision pour la classe "1" sont tous deux importants, de sorte que le score f1 pour la classe "1" doit être considéré comme la mesure la plus importante. Un score f1 élevé indique un équilibre entre l'identification du plus grand nombre possible de visiteurs potentiels (rappel élevé) et la minimisation du nombre de faux positifs (précision élevée). C'est important pour le site, car il souhaite augmenter le taux de conversion des visiteurs tout en réduisant le coût de la campagne de marketing.

````python
def metrics_calculator(clf, X_test, y_test, model_name):
    '''
    This function calculates all desired performance metrics for a given model on test data.
    '''
    y_pred = clf.predict(X_test)
    result = pd.DataFrame(data=[accuracy_score(y_test, y_pred),
                                precision_score(y_test, y_pred, average='binary'),
                                recall_score(y_test, y_pred, average='binary'),
                                f1_score(y_test, y_pred, average='binary'),
                                roc_auc_score(y_test, clf.predict_proba(X_test)[::,1])],
                          index=['Accuracy','Precision','Recall','F1-score','AUC'],
                          columns = [model_name])
    
    result = (result * 100).round(2).astype(str) + '%'                            
    return result
````

````python
def model_evaluation(clf, X_train, X_test, y_train, y_test, model_name):
    '''
    This function provides a complete report of the model's performance including classification reports, 
    confusion matrix and ROC curve.
    '''
    sns.set(font_scale=1.2)
    
    
    # Generate classification report for training set
    y_pred_train = clf.predict(X_train)
    print("\n\t  Classification report for training set")
    print("-"*55)
    print(classification_report(y_train, y_pred_train))

    # Generate classification report for test set
    y_pred_test = clf.predict(X_test)
    print("\n\t   Classification report for test set")
    print("-"*55)
    print(classification_report(y_test, y_pred_test))
    
    # Create figure and subplots 
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5), dpi=100, gridspec_kw={'width_ratios': [2, 2, 1]})
    
    # Plot confusion matrix for test set
    ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test, colorbar=False, cmap= 'Blues', ax=ax1)
    ax1.set_title('Confusion Matrix for Test Data')                                     
    ax1.grid(False)
    
    # Plot ROC curve for test data and display AUC score 
    RocCurveDisplay.from_estimator(clf, X_test, y_test, ax=ax2)
    ax2.set_xlabel('False Positive Rate')
    ax2.set_ylabel('True Positive Rate')
    ax2.set_title('ROC Curve for Test Data (Positive label: 1)')
    
    # Report results for the class specified by positive label
    result = metrics_calculator(clf, X_test, y_test, model_name)
    table = ax3.table(cellText=result.values, colLabels=result.columns, rowLabels=result.index, loc='center')
    table.scale(0.6, 2)
    table.set_fontsize(12)
    ax3.axis('tight')
    ax3.axis('off')
    # Modify color 
    for key, cell in table.get_celld().items():
        if key[0] == 0:
            cell.set_color('blue')
    plt.tight_layout()
    plt.show() 
````

````python
# Define a scorer function
def f1_metric(model, X_train, y_train):
    '''
    This function reports f1-score metric for the class specified by 'Positive Label' (or '1')
    '''
    return f1_score(y_train, model.predict(X_train), average='binary')
````

````python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
import numpy as np

# Définir les poids associés aux classes
class_weights = [{0: x, 1: 1.0 - x} for x in np.linspace(0.001, 0.5, 20)]

# Définir la grille des hyperparamètres
param_grid = [
    {'solver': ['lbfgs', 'newton-cg', 'sag', 'saga'],
     'penalty': ['none'],
     'class_weight': class_weights},

    {'solver': ['lbfgs', 'newton-cg', 'sag'],
     'penalty': ['l2'],
     'C': np.logspace(-5, 5, 10),
     'class_weight': class_weights},

    {'solver': ['liblinear', 'saga'],
     'penalty': ['l1', 'l2'],
     'C': np.logspace(-5, 5, 10),
     'class_weight': class_weights},

    {'solver': ['saga'],
     'penalty': ['elasticnet'],
     'C': np.logspace(-5, 5, 10),
     'l1_ratio': np.arange(0, 1.1, 0.1),
     'class_weight': class_weights}
]
````

````python
def tune_clf_hyperparameters(clf, param_grid, X_train, y_train):
    '''
    Cette fonction optimise les hyperparamètres pour un classifieur en recherchant sur une grille d'hyperparamètres spécifiée.
    '''
    # Créer l'objet de validation croisée en utilisant StratifiedKFold pour garantir que la distribution de classe est la même dans toutes les folds
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

    # Créer l'objet GridSearchCV
    clf_grid = GridSearchCV(clf, param_grid, cv=cv, scoring='f1', n_jobs=-1)

    # Adapter l'objet GridSearchCV aux données d'entraînement
    clf_grid.fit(X_train, y_train)

    # Obtenir les meilleurs hyperparamètres
    print("Meilleurs hyperparamètres:\n", clf_grid.best_params_)

    # Retourner le meilleur estimateur qui donne le meilleur modèle ajusté aux données d'entraînement
    return clf_grid.best_estimator_
````
